EJERCICIO 1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.


```python
# Definimos la cadena de texto
texto = "Este es mi primer proyecto de Python"

# Definimos la función
def contar_letras(texto):
    """
    Función que recibe una cadena de texto y devuelve un diccionario
    con la frecuencia de cada letra (ignorando espacios).
    """
    # Definimos un diccionario vacío
    frecuencias = {}

    # Creamos un bucle for para recorrer cada letra del texto
    for letra in texto:
        if letra != " ":  # Ignoramos los espacios
            letra = letra.lower() # Convertimos a minúscula
            # Si la letra ya está dentro del diccionario, le sumamos 1
            if letra in frecuencias:
                frecuencias[letra] += 1
            # De lo contrario, la añadimos al diccionario
            else:
                frecuencias[letra] = 1
   
    # Devolvemos el diccionario completo
    return frecuencias

# Llamamos a la función y mostramos el resultado 
contar_letras(texto)
```

EJERCICIO 3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.


```python
# Definimos la lista de palabras y la palabra objetivo
palabras = ["iguana", "analista", "pera", "persiana", "casa", "rana", "taza"]
palabra_objetivo = "ana"

# Creamos la función
def nueva_lista(palabras, palabra_objetivo):
    """
    Función que toma una lista de palabras y una palabra objetivo y devuelve 
    una lista con todas las palabras de la lista original que contengan la palabra objetivo
    
    Args:
        palabras (list): lista de palabras
        palabra_objetivo (str): palabra que será buscada dentro de la lista de palabras

    Returns:
        list: lista de palabras que contienen la palabra objetivo
    """
    # Creamos una lista vacía donde se guardarán las palabras que contengan "ana"
    lista_filtrada = [] 
    
    # Recorremos cada palabra de la lista
    for palabra in palabras:
        
        # Comprobamos si la palabra objetivo está dentro de la palabra actual
        if palabra_objetivo in palabra:
            
            # Si la condición es True, añadimos la palabra a la lista 
            lista_filtrada.append(palabra)
    
    # Devolvemos la lista completa
    return lista_filtrada

# Llamamos a la función para obtener la lista de palabras que contienen "ana"
nueva_lista(palabras, palabra_objetivo)
```

EJERCICIO 5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
# Definimos una lista con las notas de un alumno
notas_alumno = [9.7, 5, 6.3, 7.1, 6, 4.5, 9.3, 8.4, 7, 5.6, 8]

# Importamos la función de la media de la librería "statistics"
from statistics import mean

# Definimos la función
def nota_final(notas_alumno, nota_aprobado=5):
    """
    Esta función calcula la media de las notas de un alumno y devuelve una tupla con: 
        - la media (redondeada a dos decimales)
        - el estado: 'Aprobado' si la media es mayor o igual a 5, 'Suspenso' en caso contrario
    """
    
    # Calculamos la media
    media = round(mean(notas_alumno), 2)
    
    # Evaluamos si la media es suficiente para aprobar
    if media >= nota_aprobado:
        return (media, 'Aprobado')
    else: 
        return (media, 'Suspenso')

# Llamamos a la función 
nota_final(notas_alumno, nota_aprobado=5)
```

EJERCICIO 6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
# Definimos la función
def factorial(n):
    """
    Esta función calcula el factorial de un número de manera recursiva
    
    Args:
    n (int): Número entero no negativo cuyo factorial se quiere calcular.

    Returns:
    int: El factorial de n, es decir, n! = n * (n-1) * (n-2) * ... * 2 * 1.

    Ejemplo:
    factorial(5) retorna 120 porque 5*4*3*2*1 = 120.
    """

    if n == 0:
        return 1 #0! == 1
    else:
        return n * factorial(n-1)

# Llamamos a la función con el número 5 
print(factorial(5))
```

EJERCICIO 10. Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una excepción personalizada y maneja el error adecuadamente.

```python
# Creamos una lista de números, en este caso vacía
numeros = []

# Importamos la función de la media
from statistics import mean

# Creamos la excepción personalizada
class ListaVaciaError(Exception):
    pass

# Definimos la función
def promedio_lista(numeros): 
    """Función que calcula la media de una lista de números.
     Lanza una excepción si la lista está vacía."""
    # Si la longitud de la lista es igual a cero
    if len(numeros) == 0:
        # Lanzamos la excepción personalizada
        raise ListaVaciaError("La lista está vacía, no tenemos datos para calcular la media")
    # Si la longitud es distita de cero, devolvemos la media de los números
    return mean(numeros)

# Intentamos calcular la media de la lista
try:    
    # Llamamos a la función y mostramos el resultado por pantalla
    print(f'El promedio de la lista de números dada es {promedio_lista(numeros)}')

# En caso que la lista esté vacía, manejamos el error
except ListaVaciaError as e:
    print(e)
```

```python
# Definimos una frase
frase = "Hoy hace calor y ha salido el sol"

# Definimos la función
def palabras_mas_largas(frase, n):
    """Función que separa una frase en palabras y devuelve
    una lista con aquellas palabras cuya longitud es mayor que n

    Args:
        frase (string): Frase que se va a analizar
        n (int): Longitud mínima que debe tener una palabra

    Returns:
        list: Lista de palabras con una longitud mayor que n
    """

    # Separamos la frase en palabras usando los espacios como separador
    palabras = frase.split()
    # Filtramos las palabras cuya longitud es mayor que n
    return list(filter(lambda palabra: len(palabra) > n, palabras))


# Llamamos a la función pasando la frase y el valor n=3
palabras_mas_largas(frase, 3)
```

EJERCICIO 28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

```python
# Definimos una lista 
frutas = ["Naranja", "Fresa", "Mango", "Pera", "Manzana", "Banana", "Piña", "Mango", "Banana"]

# Definimos la función
def primer_duplicado(frutas):
    """
    Función que busca el primer elemento duplicado de una lista dada

    Args:
        frutas (list): lista de frutas

    Returns:
        str: primer elemento duplicado
    """
    # Definimos un set vacío al cual se añadirán las frutas que vamos a ir recorriendo hasta la primera duplicada
    frutas_vistas = set()
    
    # Recorremos la frutas con un bucle for
    for fruta in frutas:
        fruta_minus = fruta.lower() # Transformamos todas las letras a minúsculas

        # Si la fruta ya está incluída, nos devuelve esa fruta y se detiene la función
        if fruta_minus in frutas_vistas:
            return fruta
        # De lo contrario, la añade al set
        frutas_vistas.add(fruta_minus)

    # Si no hay ninguna fruta repetida nos devuelve None
    return None

# Llamamos a la función
primer_duplicado(frutas)
```

EJERCICIO 29. Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el carácter '#', excepto los últimos cuatro.

```python
# Definimos una variable
numero_tarjeta = 123456789101112

# Definimos la función
def enmascarar(numero_tarjeta):
    """
    Función que convierte una variable en cadena y 
    enmascara todos los caracteres excepto los últimos cuatro con '#'
    """
    texto = str(numero_tarjeta) # Convertimos la variable a texto
    longitud = len(texto) # Calculamos la longitud del texto
    # Si tiene 4 o menos, no enmascarar
    if longitud <= 4:
        return texto              
    # De lo contrario, enmascarar todos los caracteres excepto los últimos 4 y mostrar los últimos 4
    return '#' * (longitud - 4) + texto[-4:]

# Llamamos a la función
enmascarar(numero_tarjeta)
```

EJERCICIO 31. Crea una función que solicite al usuario ingresar una lista de nombres y luego solicite un nombre para buscar en esa lista. Si el nombre está en la lista, se imprime un mensaje indicando que fue encontrado, de lo contrario, se lanza una excepción.

```python
# Definimos la función
def buscar_nombres():
    """
    Función que busca un nombre elegido por el usuario en
    una lista escogida por el usuario también

    Raises:
        ValueError: si el nombre elegido no se encuentra en la lista de nombres
    """
    try:
        # Pedimos al usuario que escriba una lista de nombres 
        lista_nombres = input("Escriba varios nombres separados por comas, por favor")
        # Quitamos espacios extra y ponemos la primera letra de cada nombre en mayúscula
        nombres = [n.strip().capitalize() for n in lista_nombres.split(",")] 
        # Mostramos la lista de nombres transformada
        print(f'La lista de nombres es: {nombres}')
        # Pedimos al usuario que ingrese el nombre que quiere buscar
        nombre_elegido= input("Escriba un nombre (puede estar o no en la lista)").capitalize()
        
        # Comprobamos si el nombre elegido está en la lista
        if nombre_elegido.capitalize() in nombres:
             print(f'El nombre elegido "{nombre_elegido}" se ha encontrado en la lista')
        else:
            # Si no está, lanzamos un error
            raise ValueError(f'Error: El nombre elegido "{nombre_elegido}" no está en la lista')
    
    # Capturamos la excepción y mostramos el mensaje de error
    except ValueError as x:
        print(x)

# Llamamos a la función
buscar_nombres()
```

```python
# Creamos una lista de empleados y el nombre a buscar
lista_empleados = [
    {'nombre': "Juan Pérez", 'puesto': "marketing"},
    {'nombre': "Adrián López", 'puesto': "IT"},
    {'nombre': "Cristina Gómez", 'puesto': "RRHH"}
]
nombre_empleado = "Samuel López"

# Definimos la función
def buscar_empleado(nombre_empleado, lista_empleados):
    """
    Función que busca un nombre completo en la lista de empleados y 
    devuelve el puesto del empleado.
    Si el empleado no está en la lista,
    indica que esa persona no trabaja en la empresa.

    Args:
        nombre_empleado (string): nombre completo del empleado (nombre y apellido)
        lista_empleados (list): lista de diccionarios con los campos 'nombre' y 'puesto'
    
    Returns:
        str: mensaje indicando el puesto del empleado o que no trabaja en la empresa
    """

    # Limpiamos espacios extra y pasamos a minúsculas para ignorar mayúsculas y espacios extra
    nombre_transf = nombre_empleado.strip().lower()
    
    # Recorremos la lista de empleados
    for empleado in lista_empleados:
        if nombre_transf == empleado['nombre'].strip().lower():
            # Si encontramos el nombre, devolvemos el puesto
            return f"{empleado['nombre']} trabaja en la empresa en {empleado['puesto']}"
    
    # Si no encontramos el nombre, indicamos que la persona no trabaja en la empresa
    return f"{nombre_empleado.strip()} no trabaja en la empresa"

# Llamamos a la función 
buscar_empleado(nombre_empleado, lista_empleados)
```

EJERCICIO 34. Crea la clase `Arbol`, define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son: `crecer_tronco`, `nueva_rama`, `crecer_ramas`, `quitar_rama` e `info_arbol`. El objetivo es implementar estos métodos para manipular la estructura del árbol.

**Código a seguir:**
1. Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
2. Implementar el método `crecer_tronco` para aumentar la longitud del tronco en una unidad.
3. Implementar el método `nueva_rama` para agregar una nueva rama de longitud 1 a la lista de ramas.
4. Implementar el método `crecer_ramas` para aumentar en una unidad la longitud de todas las ramas existentes.
5. Implementar el método `quitar_rama` para eliminar una rama en una posición específica.
6. Implementar el método `info_arbol` para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las mismas.

**Caso de uso:**
1. Crear un árbol.
2. Hacer crecer el tronco del árbol una unidad.
3. Añadir una nueva rama al árbol.
4. Hacer crecer todas las ramas del árbol una unidad.
5. Añadir dos nuevas ramas al árbol.
6. Retirar la rama situada en la posición 2.
7. Obtener información sobre el árbol.

**Creación de la clase arbol**

```python
class Arbol:
    """
    Clase que representa un árbol

    Atributos:
        tronco (int): longitud del tronco
        ramas (list): lista de ramas, cada una representada como un diccionario con nombre y longitud
    """

    # Definimos el método constructor
    def __init__(self):
        self.tronco = 1
        self.ramas = []

    # Definimos el resto de métodos
    def crecer_tronco(self):
        """Aumenta la longitud del tronco en 1"""
        self.tronco += 1
        return self.tronco

    def nueva_rama(self, nombre, longitud=1):
        """
        Añade una nueva rama al árbol

        Args:
            nombre (str): nombre de la rama
            longitud (int, float): longitud inicial de la rama (por defecto 1)
        """
        self.ramas.append({"nombre": nombre, "longitud": longitud})

    def crecer_ramas(self):
        """Aumenta la longitud de todas las ramas en 1"""
        for rama in self.ramas:
            rama["longitud"] += 1

    def quitar_rama(self, posicion):
        """
        Elimina una rama en una posición específica

        Args:
            posicion (int): índice de la rama en la lista
        """
        if 0 <= posicion < len(self.ramas):
            self.ramas.pop(posicion)

    def info_arbol(self):
        """Devuelve información del árbol: tronco y ramas con sus longitudes"""
        nombres = [rama["nombre"] for rama in self.ramas]
        longitudes = [rama["longitud"] for rama in self.ramas]
        return (f"El tronco de mi_arbol mide {self.tronco} metros, "
                f"los nombres de las ramas de mi_arbol son {nombres} "
                f"y sus longitudes son {longitudes} metros")
```

Caso de uso

```python
# Generamos un árbol 
mi_arbol = Arbol()

# 1. Hacemos crecer el tronco del árbol en 1 metro
mi_arbol.crecer_tronco()

# 2. Añadimos una rama nueva al árbol llamada "rama1" con longitud inicial 4
mi_arbol.nueva_rama("rama1", 4)

# 3. Hacemos crecer todas las ramas del árbol en 1 metro
mi_arbol.crecer_ramas()

# 4. Añadimos dos nuevas ramas al árbol
mi_arbol.nueva_rama("rama2")        # "rama2" con longitud por defecto = 1
mi_arbol.nueva_rama("rama3", 3)     # "rama3" con longitud 3

# 5. Retiramos la rama situada en la posición 2 (segunda rama, índice 1)
mi_arbol.quitar_rama(1)

# 6. Obtenemos la información completa sobre el árbol
mi_arbol.info_arbol()
```

poder hacerse.
3. Implementar el método `transferir_dinero` para realizar una transferencia desde otro usuario al usuario actual. Lanzará un error en caso de no poder hacerse.
4. Implementar el método `agregar_dinero` para agregar dinero al saldo del usuario.

**Caso de uso:**
1. Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
2. Agregar 20 unidades de saldo de "Bob".
3. Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".
4. Retirar 50 unidades de saldo a "Alicia".

**Creación de la clase**

```python
class UsuarioBanco:

    # Creamos el método constructor y los métodos de gestión bancaria
    def __init__(self, nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    def retirar_dinero(self, cantidad):
        if cantidad <= 0:
            print("La cantidad debe ser positiva")
        elif self.saldo >= cantidad:
            self.saldo -= cantidad
            print(f'{self.nombre} ha retirado {cantidad}€, el saldo actual en la cuenta es {self.saldo}€')
        else:
            raise ValueError(f'Fondos insuficientes. El saldo actual es {self.saldo}€ y la cantidad solicitada es {cantidad}€')

    def transferir_dinero(self, cantidad, destinatario):
        if cantidad <= 0:
            print("La cantidad debe ser positiva")
        elif self.saldo >= cantidad:
            self.saldo -= cantidad
            destinatario.saldo += cantidad
            print(
            f'{self.nombre} ha transferido {cantidad}€ a {destinatario.nombre}. '
            f'El saldo actual de {self.nombre} es {self.saldo}€ y '
            f'el saldo actual de {destinatario.nombre} es {destinatario.saldo}€'
            )
        else:
            raise ValueError(f'Fondos insuficientes. El saldo disponible es {self.saldo}€')

    def agregar_dinero(self, cantidad):
        if cantidad > 0:
            self.saldo += cantidad
            print(f'{self.nombre} ha ingresado {cantidad}€. El saldo actual es {self.saldo}€')
        else:
            print('La cantidad debe ser superior a 0')
```

Casos de uso

```python
# Creamos los usuarios 
Alicia = UsuarioBanco("Alicia", 100, True)
Bob = UsuarioBanco("Bob", 50, True)

# 1. Agregar 20€ a la cuenta de Bob
Bob.agregar_dinero(20)

# 2. Hacer una transferencia de 80€ desde Bob a Alicia
try:
    Bob.transferir_dinero(80, Alicia)
except ValueError as e:
    print(e)

# 3. Retirar 50€ de la cuenta de Alicia
try:
    Alicia.retirar_dinero(50)
except ValueError as e:
    print(e)
```

EJERCICIO 36. Crea una función llamada `procesar_texto` que procesa un texto según la opción especificada: `contar_palabras`, `reemplazar_palabras`, `eliminar_palabra`. Estas opciones son otras funciones que tenemos que definir primero y llamar dentro de la función `procesar_texto`.

**Código a seguir:**
1. Crear una función `contar_palabras` para contar el número de veces que aparece cada palabra en el texto. Tiene que devolver un diccionario.
2. Crear una función `reemplazar_palabras` para remplazar una `palabra_original` del texto por una `palabra_nueva`. Tiene que devolver el texto con el remplazo de palabras.
3. Crear una función `eliminar_palabra` para eliminar una palabra del texto. Tiene que devolver el texto con la palabra eliminada.
4. Crear la función `procesar_texto` que tome un texto, una opción(entre "contar", "reemplazar" "eliminar") y un número de argumentos variable según la opción indicada.

 **Caso de uso:**
Comprueba el funcionamiento completo de la función `procesar_texto`.

   - **Paso 1**: función `contar_palabras`

```python
# Definimos la función contar_palabras
def contar_palabras(texto):
    """
    Función que cuenta cuántas veces aparece cada palabra en un texto

    Args:
        texto (str): texto a analizar

    Returns:
        dict: diccionario con cada palabra y su cantidad de apariciones
    """
    
    # Pasamos todas las palabras a minúscula y separamos por espacios
    palabras = texto.lower().split()
    
    # Creamos el diccionario vacío donde se añadirán las palabras y la cantidad de apariciones
    conteo = {}
    
    # Recorremos cada palabra
    for palabra in palabras:
        # Si la palabra ya está en el diccionario, incrementamos el contador
        if palabra in conteo:
            conteo[palabra] += 1
        # De lo contrario, añadimos la palabra al diccionario y el contador a 1
        else:
            conteo[palabra] = 1

    # Devolvemos el diccionario completo
    return conteo 
```

   - **Paso 2**: función `reemplazar_palabras`

```python
# Definimos la función
def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    """
    Función que reemplaza todas las apariciones de una palabra en un texto por otra palabra

    Args:
        texto (str): texto donde se hará el reemplazo
        palabra_original (str): palabra que se desea reemplazar
        palabra_nueva (str): palabra que reemplazará a la original

    Returns:
        str: texto con las palabras reemplazadas
    """
    return texto.replace(palabra_original, palabra_nueva)
```

   - **Paso 3**: función `eliminar_palabras`

```python
# Definimos la función
def eliminar_palabras(texto, palabra_eliminada):
    """
    Función que elimina la primera aparición de una palabra específica en un texto

    Args:
        texto (str): texto original del que se eliminará la palabra
        palabra_eliminada (str): palabra que se desea eliminar

    Returns:
        str: texto resultante después de haber eliminado la palabra
    """
    # Separamos el texto en una lista de palabras usando los espacios como separadores
    palabras = texto.split()  

    # Recorremos la lista de palabras usando enumerate para obtener tanto el índice (i) como la palabra (p) en cada iteración
    for i, p in enumerate(palabras):
        # Si encontramos la palabra que queremos eliminar
        if p == palabra_eliminada:
            # Eliminamos esa palabra usando pop con el índice correspondiente
            palabras.pop(i)  # Esto elimina solo la primera coincidencia
            break # Salimos del bucle 
    
    # Unimos la lista de palabras de nuevo en una cadena, separadas por espacios
    return " ".join(palabras)
```

   - **Paso 4**: función `procesar_texto`

```python
# Definimos la función
def procesar_texto(texto, opcion, *args):
    """
    Procesa un texto según la opción indicada: contar palabras, reemplazar o eliminar

    Args:
        texto (str): texto a procesar
        opcion (str): opción de procesamiento. Puede ser:
            - "contar": cuenta cuántas veces aparece cada palabra
            - "reemplazar": reemplaza una palabra por otra
            - "eliminar": elimina la primera aparición de una palabra
        *args: argumentos adicionales según la opción
            - "contar": no requiere argumentos
            - "reemplazar": args = (palabra_original, palabra_nueva)
            - "eliminar": args = (palabra_eliminada,)

    Returns:
        dict o str: resultado según la opción seleccionada
    """
    
    if opcion == "contar":
        return contar_palabras(texto)  # Llamamos a la función contar_palabras
    
    elif opcion == "reemplazar":
        # Si los argumentos que recibe la función no son 2, lanzamos un error
        if len(args) != 2:
            raise ValueError("Para 'reemplazar' se requieren 2 argumentos: palabra_original, palabra_nueva")
        palabra_original, palabra_nueva = args
        return reemplazar_palabras(texto, palabra_original, palabra_nueva)  # Llamamos a tu función reemplazar_palabras
    
    elif opcion == "eliminar":
        # Si la función no recibe un argumento, lanzamos un error
        if len(args) != 1:
            raise ValueError("Para 'eliminar' se requiere 1 argumento: palabra_eliminada")
        palabra_eliminada = args[0]
        return eliminar_palabras(texto, palabra_eliminada)  # Llamamos a tu función eliminar_palabras
    
    else:
        # Lanzamos un error si la opción elegida no es válida
        raise ValueError("Opción no válida. Elige entre 'contar', 'reemplazar' o 'eliminar'")
```

   Ejemplos de uso

```python
texto = "Hoy hace mucho frío y mañana también hará frío"

# Contar palabras
print(procesar_texto(texto, "contar"))

# Reemplazar "frío" por "calor"
print(procesar_texto(texto, "reemplazar", "frío", "calor"))

# Eliminar la primera aparición de "mucho"
print(procesar_texto(texto, "eliminar", "mucho"))

# Opción no válida
try:
    print(procesar_texto(texto, "cambiar"))
except ValueError as x:
    print(x)
```

EJERCICIO 39. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o "triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).

```python
import math  # Para usar pi en el círculo

def calcular_area(figura, datos):
    """
    Calcula el área de una figura geométrica.
    
    Parámetros:
    figura (str): "rectángulo", "triángulo" o "círculo"
    datos (tuple): datos necesarios según la figura
                   - rectángulo: (base, altura)
                   - triángulo: (base, altura)
                   - círculo: (radio,)
    
    Returns:
    float: área de la figura
    """

    if figura == "rectángulo":
        base, altura = datos
        return base * altura
    elif figura == "triángulo":
        base, altura = datos
        return (base * altura) / 2
    elif figura == "círculo":
        radio = datos[0]
        return math.pi * radio**2
    else:
        # Lanzamos un error si la figura es no válida
        raise ValueError("Figura no válida")  


# Calculamos áreas
area_rectangulo = round(calcular_area("rectángulo", (5, 3)), 2)
area_triangulo = round(calcular_area("triángulo", (6, 4)), 2)
area_circulo =  round(calcular_area("círculo", (3,)), 2)

# Mostramos los resultados
print(f'El área del rectángulo es {area_rectangulo}')
print(f'El área del triángulo es {area_triangulo}')
print(f'El área del círculo es {area_circulo}')
```
